```python
# preprocess.py
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.signal import savgol_filter
import os

CSV = "Final_data.csv"
OUTDIR = "data_npy"
os.makedirs(OUTDIR, exist_ok=True)

# PARAMETERS
WINDOW_SIZE = 100      # number of timesteps per window (tweak: depends on sample rate)
STEP = 50              # sliding window step (50% overlap)
SMOOTH_WINDOW = 11     # for Savitzky-Golay
SMOOTH_POLY = 3

# 1) load CSV
df = pd.read_csv(CSV)
print("Loaded:", df.shape)

# choose the numeric signal columns for per-sample sequences
# We'll create multi-channel windows: [ppg_signal, heart_rate, velocity_mean, velocity_peak, avg_temp, thermal_diff]
channels = ["ppg_signal", "heart_rate", "velocity_mean", "velocity_peak", "avg_temp", "thermal_diff"]
X_cols = df[channels].copy()

# 2) smoothing
for col in channels:
    X_cols[col] = savgol_filter(X_cols[col].values, SMOOTH_WINDOW if SMOOTH_WINDOW < len(X_cols) else 3, SMOOTH_POLY)

# 3) scaling (fit scaler on full dataset)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cols)
np.save(os.path.join(OUTDIR, "scaler_mean_std.npy"), np.array([scaler.mean_, scaler.scale_]))

# 4) sliding windows
windows = []
ids = []
for start in range(0, len(X_scaled) - WINDOW_SIZE + 1, STEP):
    w = X_scaled[start:start+WINDOW_SIZE]
    windows.append(w)
    ids.append(start)  # keep an index of window start for traceability

X = np.stack(windows).astype("float32")  # shape (num_windows, window_size, num_channels)
np.save(os.path.join(OUTDIR, "X_windows.npy"), X)
np.save(os.path.join(OUTDIR, "window_ids.npy"), np.array(ids))
print("Saved windows:", X.shape)
```

```python
# generate_pseudo_labels.py
import numpy as np
import pandas as pd
# First, install tensorflow if not already installed
import sys
import subprocess
subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'tensorflow'])
# Now import from tensorflow
from tensorflow.keras.models import load_model
import os

DATADIR = "data_npy"
CSV = "Final_data.csv"

X = np.load(os.path.join(DATADIR, "X_windows.npy"))
ids = np.load(os.path.join(DATADIR, "window_ids.npy"))
auto = load_model("models/conv1d_autoencoder.h5")

# reconstructions
X_rec = auto.predict(X, batch_size=64)
mse = np.mean((X - X_rec)**2, axis=(1,2))  # per-window MSE

# choose threshold: combination of stats or percentile
mu = mse.mean(); sigma = mse.std()
threshold = mu + 2*sigma   # conservative
# alternative: threshold = np.percentile(mse, 95)

labels = (mse > threshold).astype(int)

# summarize
import matplotlib.pyplot as plt
plt.hist(mse, bins=100)
plt.axvline(threshold, color="red")
plt.title("Reconstruction error distribution")
plt.savefig("reconstruction_error_hist.png")
print("Mean, std, threshold:", mu, sigma, threshold)
print("Label distribution:", labels.sum(), "anomalies out of", len(labels))

# Map window labels back to original CSV rows (mark windows start index)
df = pd.read_csv(CSV)
# initialize all rows as 0
df["label"] = 0
# For each window flagged as anomaly, mark its rows in original df as 1
WINDOW_SIZE = X.shape[1]
for win_start, lab in zip(ids, labels):
    if lab == 1:
        df.loc[win_start:win_start+WINDOW_SIZE-1, "label"] = 1

df.to_csv("blood_blockage_dataset_pseudo_labeled.csv", index=False)
print("Saved pseudo-labeled CSV: blood_blockage_dataset_pseudo_labeled.csv")

```
